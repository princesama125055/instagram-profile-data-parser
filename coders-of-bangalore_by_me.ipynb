```python
with open("finaldata.txt", encoding ='utf-8') as f:
    data = f.read()
```

```python
chunks = data.split("\n\n")
```

```python
def parse_data(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split("\n")
    if (len(sep_chunk)) < 5:
        return None
        
    username = sep_chunk[0]
    posts = int(sep_chunk[1].split(" post")[0].replace(",",""))
    followers = float(sep_chunk[2].split(" follower")[0].replace(",","").replace("K","").replace("M",""))       
    if ("K" in sep_chunk[2]):
            followers = int(followers*1000)
            
    elif ("M" in sep_chunk[2]):
            followers = int(followers*1000000)
    else: 
            followers = int(followers)   
            
    following = float(sep_chunk[3].split(" followin")[0].replace(",","").replace("K","0").replace("M",""))
    
    if ("K" in sep_chunk[3]):
            following = int(following*1000)
    elif ("M" in sep_chunk[3]):
            following = int(following*1000000)
    else: 
            following = int(following)
            
    name = sep_chunk[4]
        
    if (len(sep_chunk)) > 6:
            type_of_page = sep_chunk[5]
            bio = "/n".join(sep_chunk[6:])
    else:
            type_of_page = "unknown"
            bio = ""
            
    return {"username":username,"posts":posts,"followers":followers,"following":following,"name":name,"type_of_page":type_of_page,"bio":bio}

    
```
