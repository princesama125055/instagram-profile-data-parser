```python
import json

# Read raw data from text file
with open("finaldata.txt", "r", encoding="utf-8") as file:
    raw_data = file.read()
```

```python
# Split profiles using blank line separator
chunks = raw_data.strip().split("\n\n")
```

```python
def convert_to_number(text):
    """
    Convert follower/following text into integer.
    Examples:
    '22.4K followers' -> 22400
    '38K followers'   -> 38000
    '1,946 posts'     -> 1946
    """
    value = text.lower().replace(",", "").strip()

    if "posts" in value:
        value = value.replace("posts", "").replace("post", "").strip()
    elif "followers" in value:
        value = value.replace("followers", "").replace("follower", "").strip()
    elif "following" in value:
        value = value.replace("following", "").strip()

    if value.endswith("k"):
        return int(float(value[:-1]) * 1000)
    elif value.endswith("m"):
        return int(float(value[:-1]) * 1000000)
    else:
        return int(float(value))
```

```python
def parse_profile(chunk):
    """
    Parse one profile block and return a dictionary.
    Returns None if data is incomplete.
    """
    lines = [line.strip() for line in chunk.split("\n") if line.strip()]

    # Minimum required lines:
    # username, posts, followers, following, name
    if len(lines) < 5:
        return None

    try:
        username = lines[0]
        posts = convert_to_number(lines[1])
        followers = convert_to_number(lines[2])
        following = convert_to_number(lines[3])
        name = lines[4]

        # Optional fields
        if len(lines) >= 6:
            type_of_page = lines[5]
        else:
            type_of_page = "Unknown"

        if len(lines) >= 7:
            bio = "\n".join(lines[6:])   # real newline
        else:
            bio = ""

        return {
            "username": username,
            "posts": posts,
            "followers": followers,
            "following": following,
            "name": name,
            "type_of_page": type_of_page,
            "bio": bio
        }

    except Exception as e:
        print(f"Error parsing profile: {lines}")
        print(f"Reason: {e}")
        return None
```

```python
# Parse all profiles
all_profiles = []

for chunk in chunks:
    profile = parse_profile(chunk)
    if profile is not None:
        all_profiles.append(profile)

print("Total parsed profiles:", len(all_profiles))
print(all_profiles[0])
```

```python
# Save parsed data into JSON file
with open("data.json", "w", encoding="utf-8") as file:
    json.dump(all_profiles, file, indent=4, ensure_ascii=False)

print("Data saved successfully in data.json")
```
